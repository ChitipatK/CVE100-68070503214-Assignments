Question 1: Market Basket - 10 points

```python
class Basket:
    items = []
    def __init__(self, name, items):
        self.name = name
        self.items = items
    def add(self, item):
        self.items.append(item)
    def __repr__(self):
        return f"{self.name}:{self.items}"
```

```python
a = Basket("A", [])
b = Basket("B", [])
a.add("apple")
b.add("banana")
print(a, b)  # Expected Output: A:['apple'] B:['banana']
```

Question 2: Calculation Shipping Cost - 10 points

```python
def net_price(price, tax_rate, shipping):
    """Compute price + tax + shipping"""
    subtotal = price + (price * tax_rate)
    total = subtotal + shipping
    return total
print(net_price(100, 0.07, 15))
```

Question 3: Rectangle Area Calculation - 10 points

```python
def rectangle_info(width, height):
    """Return perimeter and area."""
    perimeter = 2 * (width + height)
    area = width * height
    return perimeter, area
```

```python
p = rectangle_info(height=4, width=5)
print(p)  # Expected Output: 18 20
```

Question 4: Average Score. - 10 points

```python
def average_by_key(rows):
    total = 0
    for r in rows:
        total += r["score"]
    return total / len(rows)

print(average_by_key([{"score": 5}, {"score": 7}, {"x": 1}]))  # Expected Output: 6.0
print(average_by_key([]))                                      # Expected Output: 0.0
```

Question 5: File Reader - 15 points

```python
class open_read:
    def __init__(self, path):
        self.path = path
        self.f = None
    def __enter__(self):
        self.f = open(self.path, 'r')
        return self.f
    def __exit__(self, exc_type, exc, tb):
        if self.f and not self.f.closed:
            self.f.close()
        return True
```

```python
text_path = "test_file.txt" #should create file in your computer
with open(text_path, "w", encoding="utf-8") as w:
    w.write("KMUTT\nCPE\n")
```

```python
with open_read(text_path) as fh:
    data = fh.read()
print(data.splitlines())  # Expected Output ['KMUTT','CPE']
```

Question 6: Vehicle - 20 points

```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.__score = score #private attribute, do not change this.

    def add_score(self, points):
        self.__score = self.get_score() + points

    def get_score(self):
        return self.__score

s1 = Student("John", 35)
s2 = Student("Jane", 70)
s1.add_score(20)
s2.add_score(20)
print("John score=", s1.get_score()) # Expected Output: 55
print("Jane score=", s2.get_score()) # Expected Output: 90
```
